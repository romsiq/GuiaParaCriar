## Coletando dados sobre Filmes - www.imdb.com

Instalando as bibliotecas necessárias!

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
# Função para validar a requisição de acesso ao website.

def request_webpage(url):
  res = requests.get(url)
  try:
    res.raise_for_status()
  except Exception as exc:
    print('Houve um problema com a requisição: %s' % (exc))
  return res
```

```python
# Variável para armazenar o link do website que iremos realizar o webScraping, 

coming_soon_page = request_webpage('https://www.imdb.com/movies-coming-soon/2020-07/')
```

```python
# Apresentar os dados coletados
coming_soon_page.text
```

A função a seguir chamada prettify() irá apresentar estes dados de uma forma melhor estruturada para facilitar a leitura. 

```python
coming_soon_soup = BeautifulSoup(coming_soon_page.text)
print(coming_soon_soup.prettify())
```

A seguir, vamos encontrar as imagens sobre os filmes.

Você pode localizar um elemento HTML clicando com o botão direito do mouse na página da web e selecionando 'inspecionar'. 

O código abaixo está procurando por um elemento como este:

< div class="list detail" >... (content)...< /div >

```python
details = coming_soon_soup.find('div', attrs = {'class': 'list detail'})
image_details = details.find_all('img')

#Esta lista obterá todos os dados 'src' (urls) dos pôsteres.


image_list_url = [x['src'] for x in image_details if 'poster' in x['class']]
image_list_url
```

```python
details = coming_soon_soup.find('div', attrs = {'class': 'list detail'})
image_details = details.find_all('img')

#Esta lista obterá todos os dados 'alt' (Nomes dos Filmes) próximos lançamentos.


image_list_names = [x['alt'] for x in image_details if 'poster' in x['class']]
image_list_names
```
